# Importing The Dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

# Data Collection & Processing

```python
# loading the data from sklearn
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
# Loading the data to aa data frame

data_frame = pd.DataFrame(breast_cancer_dataset.data,columns = breast_cancer_dataset.feature_names)
```

```python
# Print the first 5 rows of the dataframe

data_frame.head()
```

```python
# adding the 'target' column to the data frame

data_frame ['label'] = breast_cancer_dataset.target
```

```python
data_frame.tail()
```

```python
# number of rows and columns in the dataset

data_frame.shape
```

```python
# getting of rows and columns in the dataset

data_frame.info()
```

```python
# checking the Null Values

data_frame.isnull().sum()
```

```python
# Statistical measures about the data

data_frame.describe()
```

```python
# Checking the distribution of Target Variable

data_frame['label'].value_counts()
```

1 --> Benign

0 --> Malignant

```python
data_frame.groupby('label').mean()
```

# Separating the features and target

```python
X = data_frame.drop(columns='label' , axis=1)
Y = data_frame['label']
```

```python
print(X)
```

```python
print(Y)
```

# Splitting the data into Training data & Testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state = 2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Standardize the data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

X_train_std = scaler.fit_transform(X_train)

X_test_std = scaler.transform(X_test)
```

```python
print(X_train_std)
```

```python
# importing tensorflow and keras

import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
# Setting up the Neural Network

model = keras.Sequential([
                          keras.layers.Flatten(input_shape=(30,)),
                          keras.layers.Dense(20,activation='relu'),
                          keras.layers.Dense(2, activation='sigmoid')
])
```

```python
# compling the Neural Network

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

dog - 0, cat - 1, horse- 2


dog- [1 0 0]

cat - [0 1 0]

horse - [0 0 1]

```python
# training the Neural Network

history = model.fit(X_train_std, Y_train,  validation_split= 0.1, epochs=10)
```

Visualizing accuracy and loss

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])


plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])


plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'lower right')
```

Accuracy of the model on test data

```python
loss, accuracy = model.evaluate (X_test_std, Y_test)
print(accuracy)
```

```python
# Converting the prediction probability to class labels

Y_pred_labels = [np.argmax(i) for i in Y_pred]
print(Y_pred_labels)
```

Building the predictive system

```python
input_data = (10.8,9.71,68.77,357.6,0.09594,0.05736,0.02531,0.01698,0.1381,0.064,0.1728,0.4064,1.126,11.48,0.007809,0.009816,0.01099,0.005344,0.01254,0.00212,11.6,12.02,73.66,414,0.1436,0.1257,0.1047,0.04603,0.209,0.07699
)
# change the input_data to a numpy array
input_data_as_numpy_array = np.asarray(input_data)

# reshape the numpy array as we are predicting for one data point
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

# standardizing the input data
input_data_std = scaler.transform(input_data_reshaped)

prediction = model.predict(input_data_std)
print(prediction)

prediction_label = [np.argmax(prediction)]
print(prediction_label)

if(prediction_label[0] == 0):
  print('The tumor is Malignant')

else:
  print('The tumor is benign')
```
